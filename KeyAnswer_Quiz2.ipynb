**Question 1 (20 points)**

Please train a Perceptron model to solve a multiclass problem on a dataset with 1000 samples, 20 features, 5 classes. 
hint: you can use make_classification from sklearn to generate the dummy dataset.



```python
#import necessary libraries 
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam, SGD
from sklearn.datasets import make_classification
from tensorflow.keras.metrics import Accuracy
from sklearn.datasets import make_moons
from keras.utils import to_categorical
import keras
import numpy as np
from matplotlib import pyplot
```

```python
# 1a. create your dataset
X, y = make_classification(
    n_samples=1000,  # row number
    n_features=20, # feature numbers
    n_informative=6, # The number of informative features
    n_redundant = 2, # The number of redundant features
    n_repeated = 2, # The number of duplicated features
    n_classes = 5, # The number of classes 
    n_clusters_per_class=1,#The number of clusters per class
    random_state = 42 # random seed 
)

# 1b. Create a your perceptron layer
model = keras.Sequential([
    layers.Dense(units=5, input_shape=[20] #<- this should be same as the number of features of the dataset
                )])

model.compile(loss='sparse_categorical_crossentropy', optimizer = Adam(learning_rate=0.1), metrics='accuracy')

# 1c. Train your model
history = model.fit(
    X,
    y,
    batch_size=64,
    epochs=10,
    validation_split=0.33,
    verbose=True
)

# 1d. Show the accuracy and loss graph of your training process

pyplot.title('Loss / Mean Squared Error')
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()

# 1e. Show the weight and bias of your model
```

```python
from keras.datasets import cifar10

from keras.utils import to_categorical

from keras import layers
from keras import models

(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()  

print(train_images.shape) 
print(test_images.shape)
    
train_images = train_images.reshape((50000, 32, 32, 3))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 32, 32, 3))
test_images = test_images.astype('float32') / 255

print(train_images.shape) 
print(test_images.shape)

print(train_labels) 
print(train_labels)

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

print(train_labels) 
print(train_labels)

batch_size = 100
epochs = 2

model = models.Sequential()

model.add(layers.Conv2D(32,(5,5),activation='relu', input_shape=(32,32,3)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (5,5), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())
model.add(layers.Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])


#model.summary()
#model.fit(train_images, train_labels,
#          batch_size=batch_size,
#          epochs=epochs,
#          verbose=1)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)

print('Test loss:', test_loss)
print('Test accuracy:', test_acc)
```
